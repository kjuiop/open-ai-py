### 생성형 AI 란?

- 생성형 AI 는 텍스트, 이미지, 오디오, 합성 데이터 등 다양한 유형의 콘텐츠를 생성할 수 있는 일종의 인공지능 기술이다.

- 인공지능이란 지능형 에이전틀르 만드는 것을 다루는 컴퓨터 과학의 한 분야로, 추론하고, 학습하고, 자율적으로 행동할 수 있는 시스템입니다.

- 기본적으로 AI 는 인간처럼 생각하고 행동하는 기계를 만드는 이론과 방법과 기능이 있습니다.

<br />

### 머신러닝이란?

- 머신러닝은 AI 의 하위분야로, 입력 데이터로부터 모델을 훈련하는 프로그램이나 시스템을 말합니다.

- 훈련된 모델은 모델을 훈련하는 데 사용된 데이터에서 추출한, 이전에 본적이 없는 새로운 데이터를 통해 유용한 예측을 내릴 수 있습니다.

- 머신러닝은 컴퓨터가 명시적인 프로그래밍 없이도 학습할 수 있는 능력을 제공한다는 의미입니다.

<br />

### 머신러닝 모델은 어떤 것일까?

- 일반적인 머신 러닝 모델은 비지도 학습 모델과 지도 학습 모델로 나눠집니다.

- 둘 사이의 주요 차이점은 지도 학습 모델에서는 레이블이 있다는 것입니다.

- 레이블이 지정된 데이터는 이름, 유형, 또는 숫자와 같은 태그가 포함된 데이터입니다.

- 레이블이 없는 데이터는 태그가 없는 데이터입니다.

- 지도 학습에서 모델은 과거의 사례를 학습하여 미래의 값을 예측합니다.

- 비지도 학습에서는 발견에 관한 것이며, 원시 데이터를 살펴보고, 그것이 자연스럽게 그룹으로 분류되는지 확인하는 것입니다.

<br />

### 지도 학습 이란?

- 지도 학습에서는 테스트 데이터 값 X 가 모델에 입력됩니다.

- 모델은 예측을 출력하고 이를 모델로 학습하는 데 사용된 학습 데이터와 비교합니다.

- 예측된 테스트 데이터 값과 실제 훈련 데이터 값이 크게 다를 경우 이를 "오류"라고 합니다.

- 이 모델은 예측된 값과 실제 값이 더 가까워질 때까지 이 오류를 줄이려고 시도합니다.

- 이 방법은 고전적인 최적화 문제입니다.


<br />

### 딥 러닝이란?

- 딥러닝은 인공 신경망을 사용하여 머신 러닝보다 더 복잡한 패턴을 처리할 수 있는 머신러닝의 한 유형입니다.

- 여러 개의 상호 연결된 노드 또는 뉴런으로 구성되어 있으며, 데이터를 처리하고 예측을 통해 작업을 수행하는 방법을 학습할 수 있습니다.

- 딥 러닝 모델은 일반적으로 여러 층의 뉴런을 가지고 있어 기존 머신 러닝 모델보다 더 복잡한 패턴을 학습할 수 있습니다.

- 신경망은 레이블이 지정된 데이터와 레이블이 지정되지 않은 데이터를 모두 사용할 수 있습니다.

- 이것을 반지도 학습이라고 합니다.

- 반지도 학습에서 신경망은 적은 양의 레이블이 지정된 데이터와 많은 양의 레이블이 지정되지 않은 데이터를 사용하여 훈련됩니다.

- 레이블이 지정된 데이터는 신경망이 작업의 기본 개념을 학습하는 데 도움이 되고, 레이블이 지정되지 않은 데이터는 신경망이 새로운 예로 일반화하는 데 도움이 됩니다.

- Gen AI 는 딥 러닝의 하위 집합입니다. 

- 즉, 인공 신경망을 사용하고 지도 학습, 비지도 학습, 반지도 학습 방법을 사용하여 레이블이 지정된 데이터와 레이블이 지정되지 않은 데이터를 모두 처리할 수 있습니다.

- 대규모 언어 모델도 딥 러닝의 하위 집합입니다.

### 생성 AI 란?

- 생성형 AI 는 딥 러닝의 하위 집합입니다.

- 인공 신경망을 사용하고 지도 학습, 비지도 학습, 반지도 학습 방법을 사용하여 레이블이 지정된 데이터와 레이블이 지정되지 않은 데이터를 모두 처리합니다.

- 대규모 언어 모델도 딥 러닝의 하위 집합입니다.

- 딥 러닝 모델이나 머신 러닝 모델은 일반적으로 생성형과 판별형의 두 가지 유형으로 나눌 수 있습니다.

- 판별 모델은 데이터 포인트에 대한 레이블을 분류하거나 예측하는데 사용되는 모델 유형입니다.

- 판별 모델은 일반적으로 레이블이 지정된 데이터 포인트의 데이터 집합을 사용하여 학습되며, 데이터 포인트의 특징과 레이블 간의 관계를 학습합니다.

- 판별 모델이 훈련되며 새로운 데이터 포인트에 대한 레이블을 예측하는데 사용할 수 있습니다.

- 생성 모델은 기존 데이터의 학습된 확률 분포를 기반으로 새로운 데이터 인스턴스를 생성합니다.

- 생성 모델은 새로운 콘텐츠를 생성합니다.

<br />

### 판별 모델과 생성 모델

- 판별 모델은 조건부 확률 분포, 즉 입력 X 가 주어졌을 때 출력 Y 의 확률을 학습합니다.

- 생성 모델은 X와 Y의 확률인 결합 확률 분포, XY 의 확률 P를 학습합니다.

- 생성 모델은 새로운 데이터 인스턴스를 생성할 수 있고, 판별 모델은 서로 다른 종류의 데이터 인스턴스를 판별합니다.

<br />

### 생성형 AI 판별 기준

- 출력 또는 Y, 즉 레이블이 숫자나 클래스인 경우 생성형 AI 가 아니다.

- 출력물이 음성이나 텍스트, 오디오 또는 이미지와 같은 자연어인 경우 생성형 AI 이다.

- ```y = f(x) ``` 에서 Y는 모델 출력을 의미하고, F 는 함수, X 는 수식에 사용된 입력을 나타냅니다.

- 따라서 모든 출력은 모든 입력의 함수입니다.

- 생성형 AI 는 출력물인 Y가 판매량과 같은 숫자가 아닌 텍스트, 오디오, 이미지와 같은 자연어입니다.

- 우리가 사용하는 대응 방식은 모델이 이미 훈련된 방대한 대용량 데이터를 기반으로 합니다.

- 따라서 기존의 ML 지도 학습 과정에서는 훈련 코드와 레이블이 지정된 데이터를 사용하여 모델을 구축합니다.

- 사용 사례나 문제에 따라 모델은 예측을 제공하거나, 무언가를 분류하거나, 무언가를 클러스터링할 수 있습니다.


- 생성형 AI 프로세스는 모든 데이터 유형의 훈련 코드, 레이블이 지정된 데이터, 레이블이 지정되지 않은 데이터를 가져와 기초 모델을 구축할 수 있습니다.

- 기초 모델은 새로운 콘텐츠를 생성할 수 있고, 텍스트, 코드, 이미지, 오디오, 비디오 등을 생성할 수 있습니다.



### 생성형 AI 란 무엇인가?

- 생성형 AI 는 기존 콘텐츠에서 학습한 내용을 토대로 새로운 콘텐츠를 만드는 일종의 인공지능입니다.

- 기존 콘텐츠에서 학습하는 과정을 훈련이라고 하며, 이를 통해 통계적 모델이 생성됩니다.

- 특정 메시지가 주어지면, Gen AI 는 통계적 모델을 사용하여 예상되는 응답이 무엇일지 예측하고, 이를 통해 새로운 콘텐츠를 생성합니다.

- 데이터의 기본 구조를 학습한 후 학습에 사용된 데이터와 유사한 새로운 샘플을 생성할 수 있습니다.

- 생성 언어 모델은 학습한 내용을 활용하여 해당 정보를 기반으로 완전히 새로운 것을 만들어낼 수 있습니다.

- 그래서 생성형 AI 라는 단어를 사용합니다.

### 생성형 AI 의 특징

- Transformer 은 2018년 자연어 처리 혁명을 일으켰습니다.

- Transformer 는 인코더와 디코더로 구성됩니다.

- 인코더는 입력 시퀀스를 인코딩하여 디코더에 전달합니다. 디코더는 관련 작업에 대한 표현을 디코딩하는 방법을 학습합니다.

- Hallucinations 는 모델에 의해 생성도니 단어나 구문으로, 종종 말이 안되거나 문법적으로 틀린 경우를 말합니다.

- Hallucinations 는 데이터가 노이즈가 많거나 오염된 데이터로 훈련되었거나, 충분한 맥락이나 제약조건이 제공되지 않았을 때  발생합니다.

- Hallucinations 는 Transformer 에게 문제가 될 수 있는데, 이는 Hallucinations 으로 인해 출력된 텍스트를 이해하지 못하는 경우가 발생할 수 있기 때문입니다.

- Prompts 는 대규모 언어 모델(LLM)에 제공되는 짧은 텍스트입니다. 

- 이를 입력으로 사용할 수도 있으며, 모델의 출력을 제어하는데 사용할 수도 있습니다.

- 프롬포트 설계는 LLM 에서 원하는 결과를 생성하는 프롬포트를 만드는 과정입니다.

- 생성 AI 는 학습 데이터에 크게 의존합니다. 입력 데이터의 패턴과 구조를 분석하여 학습합니다.

### 텍스트가 입력일 때 사용할 수 있는 모델

- 텍스트 to 텍스트 : 두 텍스트 간의 매핑

- 텍스트 to 이미지 : 짧은 텍스트 설명이 포함된 많은 이미지 세트를 사용하여 확습됨

- 텍스트 to 비디오 : 텍스트 입력으로부터 비디오 표현을 생성하는 것을 목표로 함

- 텍스트 to 3D : 사용자의 텍스트 설명에 해당하는 3차원 객체를 생성하여 게임이나 다른 3D 세계에서 사용할 수 있도록 함

- 텍스트 to 테스크 : 텍스트 입력을 기반으로 정의된 작업이나 동작을 수행하도록 훈련

### Foundation 모델

- 방대한 양의 데이터를 사전 학습한 대규모 AI 모델로 감정 분석, 이미지 캡션, 객체 인식 등 다양한 다운스트림 작업에 맞게 조정 또는 미세 조정되도록 설계된 데이터

- 의료, 금융, 개인화된 고객 지원을 제공하는데 사용

- Vertex AI 는 기초 모델이 포함된 Foundation 모델을 제공함

- 언어 기반 모델에는 채팅 및 텍스트를 위한 PaLM API 가 포함됨

- 비전 기반 모델에는 텍스트 비문에서 이미지를 생성하는데 효과적

### Google Cloud 가 생성형 AI 를 활용하는 방법

1. Vertex AI Studio

- Vertex AI Studio 를 사용하면 Google Cloud 의 애플리케이션에서 활용할 수 있는 생성 AI 모델을 빠르게 탐색하고 사용자 정의할 수 있음

- 사전훈련된 모델 라이브러리, 모델 미세 조정 도구가 있음

- Vertex AI Search and Conversation 을 사용하면 고객 및 직원을 위한 생성형 AI 검색 및 대화를 구축할 수 있음

- Vertex AI 를 사용하면 자체 챗봇, 디지털 비서, 사용자 정의 검색 엔진, 지식 기반, 교육 애플리케이션 등을 만들 수 있음

<br />

2. PaLM API

- PaLM API 는 Google 의 대규모 언어 모델과 Gen AI 도구를 테스트하고 실험할 수 있음

- 개발자는 PaLM API 를 Marker Suite 와 통합하고, 그래픽 사용자 인터페이스를 통해 API 에 엑세스할 수 있음

- 이 제품 군에는 모델 학습 도구, 모델 배포 도구, 모델 모니터링 도구 등 다양한 도구가 포함되어 있음

- 모델 학습 도구는 개발자가 다양한 알고리즘을 사용하여  데이터에 대한 ML 모델을 학습하는 데 도움을 줌

- 모델 배포 도구는 개발자가 다양한 배포 옵션을 사용하여 ML 모델을 프로덕션에 배포하는 데 도움이 됨

- 모델 모니터링 도구는 개발자가 대시보드와 다양한 측정 항목을 사용하여 프로덕션에서 ML 모델의 성능을 모니터링 하는데 도움을 줌

<br />

3. Gemini AI

- 기존 언어 모델과 달리 텍스트만 이해하는데 국한되지 않고, 이미지를 분석, 오디오의 뉘앙스를 이해하고, 프로그램 코드도 해석할 수 있음
