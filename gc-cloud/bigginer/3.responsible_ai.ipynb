### Responseible AI

- AI 는 컴퓨터가 실제 세계를 이해하고 상호작용할 수 있도록 돕는다.

- AI 의 빠른 발전을 위해서는 AI 는 오류가 없어야 한다.

- 사람들은 모든 일에 개입합니다. 데이터를 수집하거나 생성하여 모델을 훈련시킵니다. AI 구현을 제어합니다. 

- 주어진 맥락에는 어떻게 적용하는지 본직적으로는 사람의 결정은 서로 얽혀 있다. 

- 사람들을 실제로 모든 결정에 자신의 가치관을 바탕으로 내린다. 

- 때문에 윤리와 책임이 중요하다. 2018년 6월 AI 의 7가지 원칙을 발표한다.

1. AI 는 반드시 사회적 차원에서 유익합니다.

2. AI 는 창조를 피해야 한다. 또는 불공정한 편견을 강화합니다. 

3. AI 는 디자인을 제공해야 합니다.

4. AI 는 사회에 책임이 있습니다. 

5. AI 는 통합해야 합니다. (개인정보 설계 원칙)

6. AI 는 높은 기준을 유지해야 합니다. 

7. AI 를 사용할 수 있어야 합니다.
