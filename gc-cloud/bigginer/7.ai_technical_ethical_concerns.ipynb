### AI technical, ethical concerns

- 윤리적 딜레마는 서로 다른 행동 방침 사이에서 어려운 선택을 해야 하는 상황, 각 행동 방침은 도덕적 원칙을 어기는 것을 수반함

- 윤리적 딜레마와 도덕적 유혹은 다르다는 것을 이해하는 것이 중요합니다.

- 도덕적 유혹은 옭고 그름 사이의 선택이며, 잘못된 일을 하는 것이 자신에게 이로운 경우입니다.

- 윤리는 가치를 표현하고 그에 따라 의문을 제기하고 정당화하는 지속적인 과정입니다.

- 권리, 의무, 사회적 이익 또는 특정 미덕과 관련된 가치에 따른 결정입니다.

- 이는 전 세계의 윤리적 틀과 이론의 접근방식이 다양하기 때문에 윤리적 심의에는 다양한 관점과 경험이 반영되는 것이 중요합니다.

### AI 윤리의 주요 우려사항

1. 투명성

- AI 시스템이 복잡해짐에 따라 사람들이 AI 시스템이 어떻게 결정을 내리는지 이해할 수 있는 만큼 충분한 투명성을 확립하는 것이 어려워질 수 있음

<br />

2. 불공정한 편견

- AI 자체적으로 불공정한 편견을 만들어내지는 않지만, 기존 사회 시스템에 존재하는 편견에 의해 증폭됩니다.

- AI 는 확장성으로 인해 불공정한 편견이 강화되고 영속성될 수 있습니다.

- 사회를 형성하는 불공정한 편견은 데이터 세트와 문제 공식화부터 모델 생성과 검증까지 AI의 모든 단계에 영향을 미칩니다.

- AI 는 설계되고 배포되는 사회적 맥락을 직접적으로 반영합니다.

- 이러한 피해를 완화하기 위해서는 사회적 맥락과 발생할 수 있는 편견을 인식하고 해결해야 합니다.

<br />

3. 보안

- AI 시스템이 사회의 중요한 구성 요소에 내장되면서 보안에 대한 중요성이 증대되고 있습니다.

- AI 의 훈련 데이터를 빼내는 것이 더욱 가치있으며, AI 를 사용하면 더 큰 규모와 속도로 공격이 가능해집니다.

- 또한 딥페이크와 같이 사람의 생체정보를 가장하는 AI 의 새로운 조작기술이 등장하고 있습니다.

<br />

4. 개인정보 보호

- AI 는 다양한 출처에서 엄청난 양의 데이터를 빠르고 쉽게 수집, 분석, 결합할 수 있는 능력을 제공합니다.

- AI 로 인해 데이터 악용, 원치않는 식별 및 추적, 침해적인 음성 및 얼굴 ㅇ니식, 프로파일링 등의 위험으로 이루어질 수 있습니다.

<br />

5. AI 사이비 과학

- AI 실무자들이 과학적 근거가 부족한 시스템을 홍보하는 경우

- 예시로 얼굴 특징을 기반으로 사람의 범죄 경향을 측정하는 AI 를 주장

<br />

6. 사람들에 대한 책임

- AI 시스템은 모든 유형의 사람들의 요구와 목표를 충족하는 동시에 적절한 인간의 지시와 제어가 가능하도록 설계되어야 합니다.

- 시스템에 대한 명확하게 정의된 목표와 운영 매개변수를 통해 다양한 방식으로 AI 시스템의 책임을 달성하도록 노력합니다.

- AI 가 언제 사용되는지에 대한 투명성, 사람들이 시스템에 개입하거나 피드백을 제공할 수 있는 능력이 필요합니다.

<br />

7. AI 로 인한 실업과 기술 저하

- 사람이 AI 에게 의존할 수록 사람의 능력이 감소할 것이라는 우려

- 그러나 주요 기술이 발전할 때마다 항상 그렇듯이 일자리가 바뀔 것으로 예상


<br />


### 생성형 AI 에만 해당하는 고유 우려 사항

1. 환각 : 생성형 AI 모델이 생성하는 콘텐츠가 비현실적이거나, 허구적이거나, 완전히 조작된 콘텐츠를 생성하는 경우를 의미합니다.

2. 사실성 : 생성형 AI 모델에서 생성된 정보의 정확성이나 진실성에 관련이 있습니다.

3. 의인화 : 기계나 AI 모델과 같은 비인간적 존재에 인관과 유사한 특성, 특징 또는 행동을 부여하는 것을 말합니다.
