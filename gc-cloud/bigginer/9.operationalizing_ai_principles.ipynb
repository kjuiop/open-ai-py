### google AI governance

- AI 원칙을 정의한 후 AI 거버넌스를 구축하는 다음 단계는 이를 실행에 옳기기 위한 검토 프로세스를 설정하는 것이다.

- 원칙은 기업이 대표하는 가치를 확립하고 기술 개발에서 무엇을 평가해야 할지 알려주는 출발점을 제공합니다.

- 이러한 AI 원칙을 적용하려면 협력적이고 지속적인 노력이 필요합니다.

<br />

### 책임있는 AI 거버넌스에 대한 몇 가지 일반적인 오해

1. 윤리적인 사람을 고용하는것이 윤리적인 AI 제품으로 보장되지 않는다.

- 윤리적인 사람조차도 윤리적 맹점에 빠질 수 있습니다. 윤리적인 의사결정을 위한 관행을 구축하고 윤리적인 심의를 위한 공간을 마련하는 것이 중요합니다.

- 윤리적인 의사결정을 위한 관행을 구축하고 윤리적인 심의를 위한 공간을 마련하는 것이 중요합니다.

<br />

2. 책임 있는 AI 에 대한 체크리스트를 만드는 것은 효과적이지 않다.

- 모든 제품의 기술적 세부 사항과 사용되는 맥락은 고유하며, 각각에 대한 평가가 필요합니다.

- 체크리스트를 따르면 비판적 사고에 한계가 생기고 윤리적 사각지대가 생길 수 있습니다.

- AI 거버넌스에 필수적인 것은 기술 검토를 지원하는 프로그램과 관행을 갖추는 것

- 특정 상황에서 모든 가능성을 상상하여 윤리적인 과제를 해결하는 것

<br />

### 검토 프로세스를 어떻게 구현하는가

1. Google 은 새로운 프로젝트, 제품 및 거래가 AI 원칙에 부합하는지 평가하기 위해 공식 검토 위원회 구조를 만듬

- AI 원칙을 구현하는 제품 영역, AI 원칙에 대한 공통된 해석을 확립하고 회사 전체에서 보정된 의사 결정을 보장

- 일상 업무와 초기평가를 담당합니다.

- 이 그룹에는 사용자 연구자, 사회과학자, 윤리학자 등 다양한 전문가가 포함되어 있어 관점과 분야의 다양성이 허용됩니다.

<br />

2. Google 전반의 다양한 분야에서 기술, 기능 및 응용 분야에 대한 전문 지식을 제공하는 고위 전문가 그룹

- 새로운 기술과 주제에 대한 전략과 지침을 제공하고, 필요한 경우 검토에 대한 컨설팅을 제공합니다.

<br />

3. 고위 임원으로 구성된 협의회

- 여러 제품과 기술에 영향을 미치는 결정을 포함하여 가장 복잡하고 어려운 문제를 처리합니다.

<br />

4. 특정 제품 영역에는 책임 있는 혁신 팀과 긴밀히 협력하는 맞춤형 AI 거버넌스 및 검토 위원회가 포함되어 있습니다.



### Google Cloud 의 검토 프로세스

- Google Cloud 의 검토프로세스는 고객 AI 거래 검토와 클라우드 AI 제품 개발 검토가 있다.

1. 고객 AI 거래 검토는 초기 단계의 고객 프로젝트를 관찰하여 일반적으로 판매되는 제품 버무이를 넘어 제안된 사용 사례가 AI 원칙과 충돌하는지 판단

2. 제품 개발 검토는 우리가 제품을 평가, 범위 지정, 구축 및 관리하는 방법에 초점을 맞춤

<br />

- 이러한 검토 프로세스는 두가지 질문에 답합니다.

1. 제안된 사용 사례가 AI 원칙과 일치하는가?

2. 의도한 이점을 실현하고 피해를 완화하기 위해 이 솔루션의 설계 및 통합에 어떤 접근 방식을 취해야 하는가?



### Celebrity Recognition case study

- AI 원칙과 검토 프로세스가 Google Cloud 의 얼굴 인식 기술 접근 방식에 어떤 영향을 미쳤는지 설명합니다.

- 2019년 Google Cloud 는 유명인 인식을 출시

- 전문 라이선스 미디어 콘텐츠에서 유명인을 태그하려는 미디어 및 엔터테인먼트 고객을 대상으로 범위가 엄격한 API 

- 기존의 편견을 강화하거나 확대하지 않아야 합니다. 특히 이것이 소외된 집단에 영향을 미칠 수 있는 경우에는 더 그렇다.

- 국제적으로 인정된 규범을 위반하는 감시에는 사용되어서는 안된다.

- 적절한 수준의 투명성과 통제를 제공하여 사람들의 개인정보를 보호해야 한다.

- 남용 가능성을 줄이고 기업 사용 사례에 맞춰 기술을 사용할 수 있도록 합니다.
