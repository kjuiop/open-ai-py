### Responsible ai apply Google Cloud

- 책임감 있는 AI 를 개발하려면 발생 가능한 문제, 한계 또는 의도치 않은 결과를 이해해야 합니다.

- 인공지능에 대해 기계가 중심적인 의사결정 역할을 한다는 오해가 널리 퍼져있다.

- 그러나 실제로 이러한 기계를 설계하고 제작하고 사용 방법을 결정하는 사람은 바로 사람이다.

- AI 개발의 모든 측면에는 사람이 참여합니다.

- 그들은 모델이 훈련되는데 사용되는 데이터를 수집하거나 생성합니다.

- 그들은 AI 의 배포와 그것이 주어진 맥락에 적용되는 방식을 제어합니다.

- 본질적으로 인간의 결정은 우리의 기술 제품 전반에 걸쳐 반영되어 있습니다.

- 그리고 사람들이 결정을 내릴 때마다 실제로는 자신의 가치에 따라 선택을 하는 것

### Google And Responsible AI

- 책임감 있는 AI 의 관힝이 없다면 겉보기엔 무해한 AI 의 사용도 여전히 윤리적 문제나 의도치 않은 결과를 야기할 수 있다.

- Google 에서는 AI 에 책임감을 부여하면 더 나은 모델을 만들고 고객과의 신뢰를 구축할 수 있다는 것을 중요하게 생각한다.

- 일련의 평가와 검토를 통해 AI 를 중심으로 제품과 사업 결정을 내린다.

- 이러한 원칙은 제품 분야와 지역 전반에 걸쳐 우리의 접근 방식에 엄격함과 일관성을 부여한다.

- 이러한 평가와 검토는 모든 프로젝트가 AI 원칙에 부합하는지 확인하는 것으로 시작된다.

### Google AI Principles

1. AI 는 사회적으로 유익해야 합니다.

2. AI 는 불공정한 편견을 만들거나 강화하는 것을 피해야 합니다.

3. AI 는 안전을 위해 제작되고 테스트되어야 하며, 해를 끼칠 수 있는 의도치 않은 결과를 피하기 위해 강력한 안전 및 보안 관행을 계속 개발하고 적용할 것

4. AI 는 사람들에게 책임을 져야 합니다. 적절한 피드백, 관련 설명 및 호소에 대한 기회를 제공하는 AI 시스템을 설계할 것입니다.

5. AI 는 개인정보 보호 설계 원칙을 통합해야 합니다.

6. AI 는 과학적 우수성에 대한 높은 기준을 유지해야 하며, 우리는 이 분야에서 과학적으로 뒷받침되는 사려 깊은 리더십을 촉진하기 위해 다양한 이해관계자와 협력할 것입니다.

7. AI 는 다양한 용도로 사용할 수 있어야 합니다. 이러한 원칙에 따르면 많은 기술에는 다양한 용도가 있으므로 잠재적으로 유해하거나 남용적인 애플리케이션을 제한하기 위해 노력해야 합니다.

### 추구하지 않는 특정 AI 응용 분야

1. 전반적인 피해를 일으키거나 일으킬 가능성이 있는 기술

2. 사람에게 부상을 입히거나 직접적으로 부상을 용이하게 하는 것을 주된 목적으로 하거나 실행하는 무기, 기타 기술

3. 국제적으로 인정된 규범을 위반하는 감시를 위해 정보를 수집하거나 사용하는 기술

4. 그 목적이 국제법과 인권의 널리 받아들여진 원칙을 위반하는 기술



### The Economist Intelligenct Unit (EIU)

- 책임 있는 AI가 조직의 핵심적인 비즈니스 고려 사항에 미칠 수 있는 영향을 종합적으로 제시

- 이 보고서의 데이터는 광범위한 데이터 기반 연구, 업계 전문가 인터뷰, 임원 설문 조사 프로그램이 출처

<br />

1. 책임 있는 AI 관행을 통합하는 것이 제품 개발에 대한 현명한 투자

- EIU 설문 응답자의 97% 는 윤리적인 AI 검토가 제품 혁신에 중요하다는데 동의

- 윤리적 검토는 새로운 기술과 관련된 잠재적인 기회와 피해를 조사하여 제품을 책임 있는 AI 설계에 더 잘 맞춤

- 데이터 세트를 면밀히 검토하고, 하위 그룹 전체의 성과를 모델링하며, 의도한 결과와 의도하지 않은 결과의 영향을 고려한다.

- AI 시스템에 대해 조직의 90% 가 윤리적인 문제에 직면했다.

- 책임 있는 AI 는 제품 이해관계자와의 신뢰를 구축하는데 중요한 요소로 사용자에게 제품의 가치와 경쟁 우위를 모두 향상시킨다.

<br />

2. 책임감 있는 AI 선구자들이 최고의 인재를 유치하고 유지한다는 것

- 뛰어난 인재는 윤리적 문제를 해결해주는 고용주에 대해 더 강한 충성심을 나타낸다.

- 공동의 약속과 책임 있는 AI 관행을 구축하는 조직은 직원과의 신뢰와 참여를 구축하는데 유리하다.

<br />

3. 데이터 보안의 중요성

- 사이버 보안과 데이터 개인정보 보호 문제가 AI 도입에 가장 커다란 장애물

- 조직에서는 데이터를 수집, 사용, 보호하는 방법에 대해 매우 신중하게 생각해야 함

- 기업 고객은 회사 자체가 자사의 데이터를 신뢰할 수 있는 호스트라는 확신을 가져야 함

- Google 에서는 공개 개인정보 보호정책을 통해 고객 데이터를 어떻게 적극적으로 보호하는지 알림

- 조직이 데이터를 신뢰할 수 있게 되면 더 크고 다양한 데이터 세트를 얻을 수 있으며, 이는 AI 성과 향상에 도움이 됨

<br />

4. AI 규제에 앞서 준비하는 것이 중요

- AI 발전에 따라 정부의 AI 규정을 시행하려는 노력이 이루어지고 있음

- AI 시스템에 대한 새로운 투명성 및 위험 관리규칙을 유럽 의회 의원에서 처음으로 지지됨

- GDPR (유럽 연합의 데이터 보호 규정) 의 예시처럼 정부의 지침을 준수하지 않는 경우 벌금 또는 사업 중단에 대한 리스크가 존재


<br />

5. 책임 있는 AI 가 수익 성장을 개선할 수 있다.

- 책임 있는 AI 는 타깃 시장 확대, 경쟁 우위 확보, 기존 고객과의 관계 개선으로 이어질 수 있다.


<br />

6. 책임 있는 AI 가 파트너십을 강화하고 있다는 점

- 투자자들은 지속 가능하고 장기적인 투자에 대한 관심이 있음

- 지속 가능한 투자에는 온실가스 배출, 다양성 이니셔티브, 임금 구조 등 ESG 요소와 위험 등을 고려함

- 책임 있는 AI 원칙에 따라 투자 대상자를 평가함

<br />

7. 강력한 신뢰와 브랜드를 유지하는 것과 관련이 있음

- 기술 기업의 신뢰는 브랜드 가치와 연결됨
