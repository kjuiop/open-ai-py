### 머닝러닝 모델의 확산모델

- Variational autoencoders (VAEs) : 이미지를 압축된 크기로 인코딩한 후 데이터 분포 자체를 학습하는 동안 디코딩하는 변형 자동 인코더

- Generative adversarial models (GANs) : 두 개의 신경망을 서로 맞붙히는데, 한 신경망은 이미지 생성기, 한 신경망은 이미지 판별기로 시간이 지남에 따라 이미지를 판별하는 능력이 향상되고, 가짜 이미지를 생성하는 능력이 향상됨

- Autoregressive models : 이미지를 픽셀 시퀀스로 처리하여 이미지를 생성하는 자동 회귀 모델

### 무조건 확산모델

- 모델에 추가 입력이나 지침이 필요없는 무조건 확산모델

- 얼굴과 같은 특정 사물의 이미지에 대해 학습하여 해당 사물의 새 이미지를 생성할 수 있음

- 반복적인 전방 확산 프로세스를 통해 구조와 데이터 분포를 파괴합니다.

- 구조와 데이터를 복원하는 역 확산 프로세스를 학습합니다.

- 모자이크 제거와 같은 사례

- 예측된 이미지와 노이즈 이미지의 차이를 최소화하도록 훈련함
